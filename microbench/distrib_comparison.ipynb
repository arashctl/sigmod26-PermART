```python
with open('setbench_zipfian_99.txt', 'r') as fp: 
    sdata = [int(l) for l in fp.readlines()]
```

```python
with open('parlay_zipfian_99.txt', 'r') as fp: 
    punique = []
    pdata = []
    for l in fp.readlines():
        if l.startswith('u'): 
            punique.append(int(l.split()[1]))
        else: 
            pdata.append(int(l)) 
```

```python
len(punique), len(pdata)
```

```python
len(sdata)
```

```python
mapping = dict()
for i, n in enumerate(punique): 
    mapping[n] = i
```

```python
pdata_processed = [mapping[n] for n in pdata]
```

```python
len(pdata_processed)
```

```python
pdata_processed[:10]
```

```python
with open('parlay_zipfian_99_processed.txt', 'w') as fp: 
    for n in pdata_processed: 
        fp.write(f'{n}\n')
    
    fp.flush()
```
